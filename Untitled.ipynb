```python
def mse(y1, y2):
  '''
  Mean square error.

  Input:
  y1 - first numpy array;
  y2 - second numpy array.

  Output a mean square error.
  '''
  return ((y1 - y2) ** 2).mean()
```

```python
x = np.array([0.1, 0.2, 0.3, 0.4])
y = np.array([650, 1000, 1500, 2200])
m = len(x)
eps = 10**(-5)
a = 3
b = 4
c = 5
d = 6
res = mse(a * np.exp(b * x + c) + d, y)
```

```python
bplus = b + eps
bminus = b - eps
resplus = mse(a * np.exp(bplus * x + c) + d, y)
resminus = mse(a * np.exp(bminus * x + c) + d, y)
db = (resplus - resminus) / (2 * eps)
print(db)
yc = a * np.exp(b * x + c) + d
db = np.sum(2/m * (yc - y) * a * x * np.exp(b * x + c))
print(db)
```
